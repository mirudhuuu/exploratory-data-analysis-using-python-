```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Convert Categorical Columns
df['International plan']=df['International plan'].map({'Yes':1,'No':0})
df['Voice mail plan']=df['Voice mail plan'].map({'Yes':1,'No':0})
df['Churn']=df['Churn'].map({True:1,False:0})
```

*Converted categorical values into numeric for easier analysis


```python
#Univariate Analysis
plt.figure(figsize=(10,5))
sns.countplot(x='Churn',data=df,palette='viridis')
plt.title("churn distribution")
plt.show()
```

*Majority of customers are non-churners
*Dataset is slightly imbalanced toward non-churn class
*churn cases are still significant for significantly

```python
#Accouunt length distribution
sns.histplot(df['Account length'],bins=30,kde=True)
plt.title("Account length distribution")
plt.show()
```

*Customers are spread across different account durations
*No strong skewness observed in account length.

```python
#Customer service calls
sns.histplot(df['Customer service calls'],bins=30,kde=True)
plt.title("Customer service calls")
plt.show()
```

*Most customers make very few service calls
*A small group has high call frequency
*High call frequency may indicate dissatisfaction.

```python
#Outlier detection
sns.boxplot(x=df['Total day minutes'])
plt.title("Total dat minutes")
plt.show()
```

*Majority of users fall within a moderate usage range
*Outliers may represent heavy users rather than errors

```python
#Customer service calls
sns.boxplot(x=df['Customer service calls'])
plt.title("Customer service calls")
plt.show()
```

*Clear outliers exist with very high call counts.
*These customers may be unhappy or facing issues.
*Strong indicator for potential churn behavior

```python
#Bivariate Analysis
sns.boxplot(x='Churn',y='Customer service calls',data=df)
plt.title("Churn vs Customer service calls")
plt.show()
```

*Churned customers tend to have more service calls
*High interaction with support correlates with dissatisfaction
*Strong relationship between service calls and churn.

```python
sns.boxplot(x='Churn',y='Total day minutes',data=df)
plt.title("Churn vs Total day minutes")
plt.show()
```

High usage customers might be sensitive to pricing or quality

```python
sns.countplot(x='International plan',hue='Churn',data=df)
plt.title("International plan vs Churn")
plt.show()
```

*Customers with international plans show higher churn rates.
*Indicates possible dissatisdaction with international services.
*Plan type significantly impacts churn.

```python
#Correlation Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.drop('State', axis=1).corr(),annot=False,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

*Strong correlation between minutes and corresponding charges.
*Customer service calls positively correlated with churn
*No extreme multicolinearity issues detected.
